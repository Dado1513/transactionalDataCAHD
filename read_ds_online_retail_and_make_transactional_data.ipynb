```python
import pandas as pd
import numpy as np
```

```python
nome_file = "Online Retail.xlsx"
df = pd.read_excel(nome_file)
```

```python
df
```

```python
df_grouped = df.groupby(['CustomerID','Description']).count() 
# product different 4223
print(df['Description'].value_counts())
print(len(df['Description'].value_counts()))
```

```python
#4372 clienti (NB molto simili al numero di prodotti differenti --> 4372)
# numero di row
row = len(df_grouped.index.levels[0])
print("Numero di righe: " +str(len(df_grouped.index.levels[0])))
# numero di colonne
col = len(df_grouped.index.levels[1])
print("Numero di colonne: "+str(len(df_grouped.index.levels[1])))
customers = list(df_grouped.index.levels[0])
items = list(df_grouped.index.levels[1])
```

```python
customers
```

```python
df_grouped
```

```python
# create a transacional data
zero_data_to_add = np.zeros(shape=(row,col))
df_transactional = pd.DataFrame(zero_data_to_add,index=customers, columns=items,dtype='uint8')
#df_transactional.loc[12346.0][20713] = 0
df_transactional
```

```python
# now create a table transzionale
for customer in customers:
    items = df_grouped.loc[customer].index
    df_transactional.loc[customer][items]=1
```
