```python
import pandas as pd
import numpy as np
from scipy.sparse.csgraph import reverse_cuthill_mckee
from scipy.sparse import csr_matrix
```

```python
# read dataframe clean (creazione da read_ds_dpp)
# datframe dimensione = 200x200
df_complete = pd.read_csv("data_complete.csv",header=None,index_col=None)
df_complete
```

```python
import matplotlib.pylab as plt
import scipy.sparse as sps
```

```python
plt.spy(df_complete)
plt.show()
```

```python
sparse = csr_matrix(df_complete)
order = reverse_cuthill_mckee(sparse)
plt.spy(df_complete.loc[list(reversed(order))])
plt.show()
```

```python
plt.spy(sparse)
plt.show()
```

```python
plt.spy(sparse[list(reversed(order))])
plt.show()
```

```python
# dataframe dopo RCM
[i,j]=np.where(df_complete.iloc[list(reversed(order))] == 1)
bw = max(i-j) +1
bw
```

```python
# dataframe inizale
[i,j]=np.where(df_complete == 1)
bw = max(i-j) +1
bw
```
